## Protein Visualization and read Residue ID 

```python
path2AB = 'hmx.B99990001.pdb' # Path to the protein complex
AB        = md.load(path2AB)
table, bonds = AB.topology.to_dataframe()              # Transform mdtraj to dataframe for simpiler analysis
# A residue 
A_num      = len(table[table['chainID']==0].serial.values)         # Get the number of residue in A
A_resid    = np.unique((table[table['chainID']==0]).resSeq.values) # A safer way to obtain residue id than np.arange(1,A_num+1)
A_xyz      = AB.xyz[:,:A_num]                                     # Get the cartesian coordinate
# B residue
B_num      = len(table[table['chainID'] == 1].serial.values)
B_resid    = np.unique((table[table['chainID'] == 1]).resSeq.values)
B_xyz      = AB.xyz[:,A_num:A_num+B_num]

# Calculating the interface residue
interface_thres = 0.5 # Atoms within 5 Angstrom of its partner are considered at the PP interface
if_interface = cdist(A_xyz[0],B_xyz[0]) < interface_thres  # Calculate the interatomic distance
A_int = np.sum(if_interface,axis=1) # Simply sum up the 1st axis to get the interface atom
B_int = np.sum(if_interface,axis=0)
A_intres = []
B_intres = []
for atomid, contact in enumerate(A_int):
    if contact:
        A_intres.append(table[table['serial']==atomid+1].resSeq.values)  # counting from 1
        
for atomid, contact in enumerate(B_int):
    if contact:
        B_intres.append(table[table['serial']==atomid+A_num+2].resSeq.values) # counting from A_num + TER
A_intres = np.unique(A_intres) # Get rid of repeating residue id
B_intres = np.unique(B_intres) # Get rid of repeating residue id
```

```python
# Visualize the interatomic distance between A and B
sns.heatmap(cdist(A_xyz[0],B_xyz[0]))
```

```python
print('Molecule A has residue', A_resid)
print('Molecule B has residue', B_resid)
print('Molecule A has interface residue', A_intres)
print('Molecule B has interface residue', B_intres)
```

```python
# Visualization
view = nv.show_mdtraj(AB)
view.clear()
view.add_cartoon(selection=":A",opacity=1,color="blue")
view.add_cartoon(selection=":B",opacity=1,color="red")
# Check if the interface residues are at the interface
for res in np.concatenate([A_intres,B_intres]):
    view.add_ball_and_stick(selection=str(res),opacity=1,aspectRatio=1)
view
```

# Pocket Analysis

## Read PDB and interface residue

```python
path2out = 'hmx.B99990001_out/hmx.B99990001_out.pdb'
pdb      = md.load(path2out)
table, bonds = pdb.topology.to_dataframe()
# A SASA
A_num      = len(table[table['chainID']==0].serial.values)       # Get the number of residue in A
A_xyz      = pdb.xyz[:,:A_num]                                   # Get the cartesian coordinate
A_atomType = table.element.values[:A_num]                        # Get the atom type
A_Radii    = [_ATOMIC_RADII[element] for element in A_atomType]  # Get the Radii of A
A_SASA     = Sasa_calc(A_xyz,A_Radii)[0]                         # Calculate SASA using the function defined in function.py

# B SASA
B_num      = len(table[table['chainID']==1].serial.values)
B_xyz      = pdb.xyz[:,A_num:A_num+B_num]
B_atomType = table[table['chainID']==1].element.values 
B_Radii    = [_ATOMIC_RADII[element] for element in B_atomType]  
B_SASA     = Sasa_calc(B_xyz,B_Radii)[0]
```

```python
table.head()
```

## Read Pocket and Calculate BSA

```python
# P_SASA
P_table  = table[table['chainID']==2]
pocket_number =  max(P_table.resSeq)
P_COM  = np.zeros([pocket_number,3])
P_SIZE = np.zeros([pocket_number,3],dtype=int)

P_SASA  = np.zeros(pocket_number)
AP_SASA = np.zeros_like(P_SASA)
BP_SASA = np.zeros_like(P_SASA)
S_Drug  = np.zeros_like(P_SASA)
S_Pocket= np.zeros_like(P_SASA)
pocket_dir = 'hmx.B99990001_out/pockets/'
P_num_total = 0
for idx in range(pocket_number):
    P_num         = len(P_table[P_table['resSeq']==idx+1].serial.values)
    P_xyz         = pdb.xyz[:,A_num+B_num+P_num_total:A_num+B_num+P_num_total+P_num]
    P_num_total  += P_num
    P_Radii       = [0.152]*P_num
    P_COM[idx],P_SIZE[idx] = pocket2boxsize(P_xyz)
    P_SASA[idx]   = Sasa_calc(P_xyz,P_Radii)[0]
    AP_SASA[idx]  = Sasa_calc(np.concatenate([A_xyz, P_xyz],axis=1),A_Radii + P_Radii)[0]
    BP_SASA[idx]  = Sasa_calc(np.concatenate([B_xyz, P_xyz],axis=1),B_Radii + P_Radii)[0]
    path2pqr = ''.join([pocket_dir,'pocket',str(idx+1),'_vert.pqr'])
    S_Pocket[idx], S_Drug[idx] = read_output(path2pqr,pocket_pattern)[0]
```

```python
df_P_SASA = pd.DataFrame(np.array([P_SASA,AP_SASA,BP_SASA,S_Drug,S_Pocket]).T,columns=['SASA$_P$','SASA$_{AP}$','SASA$_{BP}$','Drug Score','Pocket Score'])
df_P_SASA['Pocket']  =  np.arange(1,pocket_number+0.1,dtype=int)
df_P_SASA['BSA$_{AP}$'] =  df_P_SASA['SASA$_P$'] + A_SASA - df_P_SASA['SASA$_{AP}$']
df_P_SASA['BSA$_{BP}$'] =  df_P_SASA['SASA$_P$'] + B_SASA - df_P_SASA['SASA$_{BP}$']
df_P_SASA['BSA$_{min}$'] =  np.min([df_P_SASA['BSA$_{AP}$'].values, df_P_SASA['BSA$_{BP}$'].values],axis=0)
```

```python
margin_size = 5 
grid_size=0.375
df_P_SASA['COM$_x$'] = P_COM[:,0]
df_P_SASA['COM$_y$'] = P_COM[:,1]
df_P_SASA['COM$_z$'] = P_COM[:,2]
df_P_SASA['SIZE$_x$'] = (P_SIZE[:,0]+margin_size)/grid_size
df_P_SASA['SIZE$_y$'] = (P_SIZE[:,1]+margin_size)/grid_size
df_P_SASA['SIZE$_z$'] = (P_SIZE[:,2]+margin_size)/grid_size
```

## Calculate BSA

```python
output_path = '../Step3_Docking_preparation/3_pdb_output/'
pdb_paths = [''.join([output_path,path]) for path in os.listdir(output_path)]
pdb_paths = [path for path in pdb_paths if '_1.pdb' in path] # Use this if you only want the first binding mode

ligand_id = np.zeros(len(pdb_paths),dtype=int)
mode = np.zeros(len(pdb_paths),dtype=int)
score = np.zeros(len(pdb_paths))
L_SASA = np.zeros(len(pdb_paths))
AL_SASA = np.zeros(len(pdb_paths))
BL_SASA = np.zeros(len(pdb_paths))

for idx, path in enumerate(pdb_paths):
    ligand_id[idx] = int(path.split('_')[-2])
    mode[idx]     = int(path.split('_')[-1].split('.pdb')[0])
    score[idx]    = read_output(path,pattern=vina_pattern)[0]
    ligand        = md.load(path)
    L_Radii       = [_ATOMIC_RADII[atom.element.symbol] for atom in ligand.topology.atoms] 
    L_SASA[idx]   = Sasa_calc(ligand.xyz,L_Radii)[0]
    AL_SASA[idx]  = Sasa_calc(np.concatenate([A_xyz,ligand.xyz],axis=1),A_Radii + L_Radii)[0]
    BL_SASA[idx]  = Sasa_calc(np.concatenate([B_xyz, ligand.xyz],axis=1),B_Radii + L_Radii)[0]
```

```python
df_vina_SASA = pd.DataFrame(np.array([score,L_SASA,AL_SASA,BL_SASA]).T,columns=['Docking Score','SASA$_L$','SASA$_{AL}$','SASA$_{BL}$'])
df_vina_SASA['BSA$_{AL}$'] =  df_vina_SASA['SASA$_L$'] + A_SASA - df_vina_SASA['SASA$_{AL}$']
df_vina_SASA['BSA$_{BL}$'] =  df_vina_SASA['SASA$_L$'] + B_SASA - df_vina_SASA['SASA$_{BL}$']
df_vina_SASA['BSA$_{min}$'] =  np.min([df_vina_SASA['BSA$_{AL}$'].values, df_vina_SASA['BSA$_{BL}$'].values],axis=0)
df_vina_SASA['Ligand'] = ligand_id
df_vina_SASA['Mode'] = mode
```

```python
top_N = 100
top_pdb_dir = '../Step4_Virtual_screening/4_top_pdb'
lig_ID = np.zeros(top_N,dtype=int)
score  = np.zeros(top_N)
L_SASA = np.zeros(top_N)
AL_SASA = np.zeros(top_N)
BL_SASA = np.zeros(top_N)

for i,idx in enumerate(smallest[:top_N]):
    lig_ID[i]     = Ligand_idx[idx]
    pdb_path      = ''.join([top_pdb_dir,'/Ligand_',str(lig_ID[i]),'.pdb'])
    ligand        = md.load(pdb_path)
    score[i]      = min_score[idx]
    L_Radii       = [_ATOMIC_RADII[atom.element.symbol] for atom in ligand.topology.atoms] 
    L_SASA[i]   = Sasa_calc(ligand.xyz,L_Radii)[0]
    AL_SASA[i]  = Sasa_calc(np.concatenate([A_xyz,ligand.xyz],axis=1),A_Radii + L_Radii)[0]
    BL_SASA[i]  = Sasa_calc(np.concatenate([B_xyz, ligand.xyz],axis=1),B_Radii + L_Radii)[0]
```

## Energy Decomposition

```python
fig, axs = plt.subplots(3,2,figsize=(13,7))
interface_res = [A_intres, B_intres-len(A_resid)]
colors        = ['b','r']
Titles        = ['Protein A', 'Protein B']
for idx, obj in enumerate(ABL_array):
    gbsa_decomp = [obj.AL_gbsa_decomp, obj.BL_gbsa_decomp]
    for jdx, ax in enumerate(axs[idx].flat):
        avg = np.average(gbsa_decomp[jdx],axis=0)
        
        std = np.std(gbsa_decomp[jdx],axis=0)
        axs[idx][jdx].plot(avg,color=colors[jdx],lw=3)
        axs[idx][jdx].fill_between(np.arange(0,len(avg+1)),avg-std,avg+std,color=colors[jdx],alpha=0.3)
        for res in interface_res[jdx]:
            axs[idx][jdx].plot(res,0.5,'ko',ms=3)
        axs[idx][jdx].grid()
        axs[idx][jdx].set_xlabel('Residue ID')
        axs[idx][jdx].set_ylabel('$\Delta\Delta$G \n (kcal/mol)')
        axs[idx][jdx].set_title(''.join([obj.name,' (',Titles[jdx],')']))

plt.tight_layout()
plt.savefig('./Step6_decomp.jpeg')
```
