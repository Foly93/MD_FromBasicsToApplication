```python
def coulomb_potential(r):
    # for potential between two charges of magnitude 1
    return 1/r

def F1_switch(r,r_c=1):
    result = np.sign(1 - r/r_c) *  (1 - 2*(r/r_c) + (r/r_c)**2)
    result[result < 0] = 0
    return result

def F2_switch(r,r_c=1):
    result = np.sign(1 - r/r_c) *  (1 - 2*(r/r_c)**2 + (r/r_c)**4)
    result[result < 0] = 0
    return result
```

```python
import matplotlib.pyplot as plt
from matplotlib.cm import viridis
import numpy as np
```

```python
r_c = 1000

x_values = np.linspace(0.2*r_c,r_c,1000)
```

```python
reduced_potential = coulomb_potential(x_values/r_c)/ coulomb_potential(np.min(x_values/r_c))
plt.plot(x_values/r_c,reduced_potential, c=viridis(0.75), label='Reduced Coulomb Potential')
plt.plot(x_values/r_c,reduced_potential*F1_switch(x_values, r_c), linestyle='-.', c=viridis(0.25), label='Reduced F1 Coulomb Potential')
plt.plot(x_values/r_c,reduced_potential*F2_switch(x_values, r_c), linestyle='--', c=viridis(0.5), label='Reduced F2 Coulomb Potential')
plt.xlabel('r/r_c')
plt.ylabel('reduced U_c')
plt.xlim(0.2,1.0)
plt.ylim(0.0,1.0)
plt.legend()
plt.savefig('reduced_coulomb_potential.svg')
```

```python
def coulomb_force(r):
    # for potential between two charges of magnitude 1
    return 1/(r*r)

def coulomb_force_F1(r,r_c=1):
    # for potential between two charges of magnitude 1
    return (1/r**2 - 1/r_c**2)

def coulomb_force_F2(r,r_c=1):
    # for potential between two charges of magnitude 1
    return 1/r**2 + 2/r_c**2 - 3 * r**2 / r_c**4
```

```python
r_c = 1
x_values = np.linspace(0.2*r_c,r_c,1000)

reduced_force = coulomb_force(x_values/r_c) / coulomb_force(np.min(x_values/r_c))
reduced_forceF1 = coulomb_force_F1(x_values/r_c, r_c)/ coulomb_force(np.min(x_values/r_c))
reduced_forceF2 = coulomb_force_F2(x_values/r_c, r_c)/ coulomb_force(np.min(x_values/r_c))

plt.plot(x_values/r_c,reduced_force, c=viridis(0.75), label='Reduced Coulomb Force')
plt.plot(x_values/r_c,reduced_forceF1, linestyle='-.', c=viridis(0.25), label='Reduced F1 Coulomb Force')
plt.plot(x_values/r_c,reduced_forceF2, linestyle='--', c=viridis(0.5), label='Reduced F2 Coulomb Force')
plt.xlabel('r/r_c')
plt.ylabel('reduced F_c')
plt.xlim(0.2,1.0)
plt.ylim(0.0,1.1)
plt.legend()
plt.savefig('reduced_coulomb_force.svg')
```
